# ML project


## Load data


```
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

```
x=df.drop('logS',axis=1)
x
```

```
y=df['logS']
y
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)

x_train
```

```
x_test
```

# Model Building

## Linear regression

```
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```
y_lr_train_ped=lr.predict(x_train)
y_lr_test_ped=lr.predict(x_test)

print(y_lr_train_ped,y_lr_test_ped)
```

# Model performance

```
y_train
```

```
y_lr_train_ped
```

```
from sklearn.metrics import mean_squared_error,r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_ped)
lr_train_r2=r2_score(y_train,y_lr_train_ped)

lr_test_mse=mean_squared_error(y_test,y_lr_test_ped)
lr_test_r2=r2_score(y_test,y_lr_test_ped)

print("LR MSE(TRAIN):",lr_train_mse)
print("LR r2(TRAIN):",lr_train_r2)
print("LR MSE(TEST):",lr_test_mse)
print("LR r2(TEST):",lr_test_r2)
```

```
lr_result=pd.DataFrame(['LineraRegression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()

lr_result.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
lr_result
```

```
from sklearn.ensemble import  RandomForestRegressor
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

```
y_rf_train_ped=rf.predict(x_train)
y_rf_test_ped=rf.predict(x_test)

print(y_rf_train_ped,y_rf_test_ped)
```

```
from sklearn.metrics import mean_squared_error,r2_score
rf_train_mse=mean_squared_error(y_train,y_rf_train_ped)
rf_train_r2=r2_score(y_train,y_rf_train_ped)

rf_test_mse=mean_squared_error(y_test,y_rf_test_ped)
rf_test_r2=r2_score(y_test,y_rf_test_ped)



rf_result=pd.DataFrame(['ForestRegressor',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()

rf_result.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
rf_result
```

```
df_model=pd.concat([lr_result,rf_result],axis=0)
df_model
```
